```python
import os
import json
import pandas as pd
```

```python
def landcover_classification():
  print('\n-----------------------------------------------------------------------------------------------')
  print('(1) Landcover classification')
  print('-----------------------------------------------------------------------------------------------')
  land_dir = os.path.join(variable['main_dir'], 'workflow_outputs', '1_HRU_data', 'Landcover')
  if not os.path.exists(land_dir):
    os.makedirs(land_dir)

  if variable['landuse_ID'] == "NA":
    print('Please upload landcover csv file to: ', land_dir)
  else:
    landID_val = list(variable['landuse_ID'].split(","))
    landclass_val = list(variable['landuse_Classifications'].split(","))
    land_class = {
        'Landuse_ID': landID_val,
        'LAND_USE_C': landclass_val
    }
    land_class_df = pd.DataFrame(land_class)
    land_class_df.to_csv(os.path.join(land_dir,'landcover_info.csv'), index=False)
    print(land_class_df)
```

```python
def vegetation_classification():
  print('\n-----------------------------------------------------------------------------------------------')
  print('(2) Vegetation classification')
  print('-----------------------------------------------------------------------------------------------')
  veg_dir = os.path.join(variable['main_dir'], 'workflow_outputs', '1_HRU_data', 'Landcover')
  if not os.path.exists(veg_dir):
    os.makedirs(veg_dir)

  if variable['veg_ID'] == "NA":
    print('Please upload vegetation csv file to: ', veg_dir)
  else:
    vegID_val = list(variable['veg_ID'].split(","))
    vegClass_val = list(variable['veg_Classifications'].split(","))
    veg_class = {
        'Veg_ID':  vegID_val,
        'VEG_C': vegClass_val
    }

    veg_class_df = pd.DataFrame(veg_class)
    veg_class_df.to_csv(os.path.join(veg_dir,'veg_info.csv'), index=False)
    print(veg_class_df)
```

```python
def soil_classification():
  print('\n-----------------------------------------------------------------------------------------------')
  print('(3) Soil classification')
  print('-----------------------------------------------------------------------------------------------')
  soil_dir = os.path.join(variable['main_dir'], 'workflow_outputs', '1_HRU_data', 'Soil')
  if not os.path.exists(soil_dir):
    os.makedirs(soil_dir)

  if variable['soil_ID'] == "NA":
    print('Please upload soil csv file to: ', soil_dir)
  else:
    soilID_val = list(variable['soil_ID'].split(","))
    soilClass_val = list(variable['soil_Classifications'].split(","))
    soil_class = {
        'Soil_ID':  soilID_val,
        'SOIL_PROF': soilClass_val
    }

    soil_class_df = pd.DataFrame(soil_class)
    soil_class_df.to_csv(os.path.join(soil_dir,'soil_info.csv'), index=False)
    print(soil_class_df)
```
